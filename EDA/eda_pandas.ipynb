```python
# empty columns

empty = []
small = []
for col in df.columns:
    count = df.value_counts(subset=col).sum()
    print(f'{col = :20} {count = }')
    if count == 0:
        empty.append(col)
    elif count < 10:
        count_dict = {col: count}
        small.append(count_dict)

print(f'Empty columns: {empty}')
print(f'Small columns: {small}')

```

```python
# Row duplicates

dup_counts = df.duplicated().value_counts()
print(f'Duplicate row counts = \n{dup_counts}')
# print(dup_counts)

dedup_df = df.drop_duplicates()
print(f'Dedup shape = {dedup_df.shape}')

dedup_counts = dedup_df.duplicated().value_counts()
print(f'Dedup row counts = \n{dedup_counts}')
# print(dedup_counts)
```
