```python
#Import Libraries
import numpy as np
import pandas as pd
import os
from pathlib import PurePath

# eda_file = "data/microwave/MR01D-2022.10.31-11.04.csv"
# eda_file = "data/microwave/MR01D-2022.10.31-11.04x.csv"
# eda_file = "data/microwave/MR01D-2022.11.04-07.csv"
eda_file = "data/microwave/MR01D-2022.11.04-07x.csv"
# eda_file = "data/microwave/MR01D-2022.11.23-28.csv"
# eda_file = "data/microwave/MR01D-2022.11.23-28x.csv"

# p = PurePath(eda_file)

    
path = os.path.dirname(eda_file)
print(f'{path =}')
dirname = PurePath(eda_file).parent
print(f'{dirname =}')
basename = PurePath(eda_file).stem
print(f'{basename =}')
resultname = basename + '_new'
print(f'{resultname =}')
extension = PurePath(eda_file).suffix
print(f'{extension =}')

df = pd.read_csv(eda_file)

print(f'before: shape={df.shape}')
df2 = df.dropna()
print(f'after: shape={df2.shape}')

# df = df.head(40000)
# print(f'trimmed: shape={df.shape}')
```

```python
# Remove fixed columns
column_blacklist = [
'Tool ID',
'Tool Description',
'Tool Area'
]

# walk thru labels
for column_name in list(df):
    # check for match
    if column_name in column_blacklist:
        # remove column if found
        df.drop(column_name, axis=1, inplace=True)

print(f'after column drop: shape={df.shape}')

```

```python
# Normalize labels
normalize_dict = {
       'TDS: Sig: Runtime' : 'Runtime'
}

# walk thru labels
for column_name in list(df):
    # check for match
    if column_name in normalize_dict.keys():
        # modify label if found
        fixit = {column_name: normalize_dict[column_name]}
        print(f'DBG: {fixit =}')
        df = df.rename(columns=fixit)

print(f'after column drop: shape={df.shape}')
print(f'column_list={list(df)}')
```

```python
# save results
resultpath = str(dirname) + '/' + resultname + extension
print(f'{resultpath =}')

df.to_csv(resultpath, index=False)
```
